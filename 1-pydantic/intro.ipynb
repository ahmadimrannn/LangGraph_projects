```python
from pydantic import BaseModel
from typing import List
from typing import Optional 
```

```python
# Pydantic Using and Creating Models

class Person(BaseModel):
  name: str
  age: int
  address: str

person = Person(name="hello david", age="24", address="pakistan")
print(person)
```

```python
# Pydantic Using and Creating Models

class Employee(BaseModel):
  name: str
  role: str
  salary: int
  
employee = Employee(name="Ahmad Imran", role="Director of AI Agents", salary=400000)
print(employee)
```

```python
# Pydantic List

class Student(BaseModel):
  name: str
  friends: List[str]
  position: int
  desired_income: Optional[int] = None


student = Student(
  name="ahmad imran", 
  friends={"moeed", "ahmad buttt", "ahmad wakeel"}, 
  position=2, 
)

print(student)
```

```python
# Nested Pydantic Model

class Address(BaseModel):
  street: int
  address: str
  zip_code: int

class Customer(BaseModel):
  name: str
  id: str
  address: Address

customer = Customer(
  name="Ahmad Imran",
  id="H543R5",
  address=Address(
    street=45,
    address="Regent Street, California",
    zip_code=50250
  )
)

print(customer)
```

```python
# Pydantic Field Funtion: Customizations and Constraints

from pydantic import Field

class Item(BaseModel):
  item_name: str = Field(min_length=1, max_length=50)
  price: float = Field(gt=0, le=50)
  quantity: int = Field(gt=0)

item = Item(item_name="Surrounded By Idiots (Book)", price=13.99, quantity=3)
print(item)
```
